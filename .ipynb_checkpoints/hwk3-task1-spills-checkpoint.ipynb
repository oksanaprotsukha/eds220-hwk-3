```python
### ===== FINAL CODE =====

# import libraries with standard abbreviations
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import geopandas as gpd

# ======= IMPORT DATA ========

# Import the NY spills data: csv
spills = pd.read_csv('data/Spill_Incidents_20231107.csv')

# Import the US counties data: shapefile with geospatial components
counties = gpd.read_file('data/tl_2022_us_county/tl_2022_us_county.shp')

# ======= SPILLS DATA PREPARATION =======

# convert column names to lower snakecase
spills.columns = spills.columns.str.lower().str.replace(' ', '_')

# convert spill_date and received_date to date format
spills[['spill_date','received_date']] = spills[['spill_date','received_date']].apply(pd.to_datetime)

# Select data about petroleum spills for the period >= 2023-01-01 and <= 2023-10-31 
petr_23 = spills[(spills.spill_date >= '2023-01-01') 
                 & (spills.spill_date <= '2023-10-31')
                 & (spills.material_family == 'Petroleum')]

# create new dataframe with the number of petroleum spills per county
spills_per_county = petr_23.groupby('county').spill_number.count()

# reset index
spills_per_county = spills_per_county.reset_index()

# rename column spill_number to n_spills
spills_per_county = spills_per_county.rename(columns={'spill_number':'n_spills'})

# ======= COUNTIES DATA PREPARATION =======

# convert column names to lower case
counties.columns = counties.columns.str.lower()

# select all counties in the state of New York (statefp = 36)
ny_counties = counties[counties.statefp == '36']

# ======= MERGE DATASETS =======

# rename county names in spills_per_county dataframe to standard values
ny_counties = ny_counties.rename(columns={'name':'county'})

# join ny_couties and spills_per_county dataframes 
ny_counties = pd.merge(ny_counties,
                       spills_per_county,
                       how='inner',
                       on='county')

# get the index of the row where county = Niagara
niagara_index = ny_counties[ny_counties.county=='Niagara'].index[0]

# add 1 oil spill for Tuscarora Indian Reservation to Niagara county
ny_counties.at[niagara_index, 'n_spills'] += 1

# ======= CREATE MAP =======

fig, ax = plt.subplots(figsize = (10,6))

# remove axis
ax.axis('off')

ny_counties.plot(ax=ax, 
                 column='n_spills',
                 edgecolor="0.8",
                 markersize=30,
                 cmap='Reds', 
                 legend=True,
                 legend_kwds={'label':'Oil spills count', 'shrink':0.9})

ax.set_title('Oil spills in New York counties in 2023',  fontsize=20)

# annotate the data source
ax.annotate("Data: New York State Department of Environmental Conservation, accessed November 7, 2023 \nhttps://www.census.gov/geographies/mapping-files/time-series/geo/tiger-line-file.2022.html", 
            xy=(0.05, .01), # position
            #xycoords='figure fraction', 
            fontsize=10, 
            xycoords ='figure fraction',
            color='#555555')

# annotate updated datapoint
ax.annotate("* One oil spill in Tuscarora Indian Reservation is counted within Niagara County.", 
            xy=(0.1, .1), # position
            #xycoords='figure fraction', 
            fontsize=10, 
            xycoords ='figure fraction',
            color='#555555')

plt.show()
```
